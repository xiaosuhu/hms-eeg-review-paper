```python
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
### This code is just one time use to convert the parquet to npy
```

```python
# --- Configuration for EEG---
parquet_dir = "F:\\Python_Project\\HMS_EEG\\Data\\train_eegs"  # input directory
output_dir = "F:\\Python_Project\\HMS_EEG\\data_preprocessed\\eeg_npy"       # output directory
os.makedirs(output_dir, exist_ok=True)

# --- Loop through all Parquet files ---
for filename in tqdm(os.listdir(parquet_dir)):
    if filename.endswith(".parquet"):
        eeg_id = os.path.splitext(filename)[0]  # remove .parquet
        parquet_path = os.path.join(parquet_dir, filename)
        npy_path = os.path.join(output_dir, f"{eeg_id}.npy")
        
        # Load Parquet -> DataFrame -> NumPy
        df = pd.read_parquet(parquet_path)
        arr = df.to_numpy(dtype=np.float32)
        
        # Save to .npy
        np.save(npy_path, arr)

        # print(f"Saved: {npy_path} shape={arr.shape}")
```
